feature selection

```python
df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
X  = df_features['Movie_Genre']+' '+ df_features['Movie_Keywords']+' '+ df_features['Movie_Tagline']+' '+ df_features['Movie_Cast']+' '+ df_features['Movie_Director']
```

```python
X
```

```python
X.shape
```

get feature text conversion to token

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

get similarity score using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(X)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

Get movie name as input as validatted for closest spelling

```python
Favourite_Movie_Name = input("Enter a Movie name:")
```

```python
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movies_Reccomendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_Title_List)
```

```python
print(Movies_Reccomendation)
```

```python
Close_Match = Movies_Reccomendation[0]
print(Close_Match)
```

```python
Recommendation_score = list(enumerate(Similarity_Score))
print(Recommendation_score)
```
